# Dataset Description:

This is a Glass Identification Dataset, collected from USA Forensic Science Service, published in UCI https://archive.ics.uci.edu/ml/datasets/Glass+Identification.
The dataset is about classification task. The data includes 9 oxide content (e.g. Na, Fe, K, etc) attributes of 6 types of glass. The analysis of the dataset can be applied to fields such as criminological investigation.

The dataset has 214 instances in total. Each has 9 attributes/features (excluding ID) and 1 label/target as follows. The attribute datatype is Real, and the label datatype is Integer. The dataset has no missing values.

Attributes information:

    X[0]: RI (Refractive Index)

    X[1]: Na (Sodium)

    X[2]: Mg (Magnesium)

    X[3]: Al (Aluminum)

    X[4]: Si (Silicon)

    X[5]: K  (Potassium)

    X[6]: Ca (Calcium)

    X[7]: Ba (Barium)

    X[8]: Fe (Iron)

Type of glass:

    value[0]: building_windows_float_processed

    value[1]: building_windows_non_float_processed

    value[2]: vehicle_windows_float_processed

    value[3]: containers

    value[4]: tableware

    value[5]: headlamps



```python
import numpy as np
import pandas as pd

# load the dataset using pandas
data = pd.read_table('glass.txt', header=None, sep=',', index_col=0)
print(data.shape)         # show the row numbers and column numbers of the dataset
data.head(10)             # show the first 10 rows of the dataset
```

```python
from sklearn.model_selection import train_test_split

X = data.iloc[:, :9]     # extract the first 9 columns as features
y = data.iloc[:, 9]      # extract the last one column as label
#print(X)

# train-test split, 2:1 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)    # shuffle the data before splitting
#print(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz

# build and train a decision tree with most depth = 1
dt1 = DecisionTreeClassifier(max_depth=1)
dt1 = dt1.fit(X_train, y_train)

# visualize the tree using graphviz
ftr_names = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']      # specify the names of the features and labels
lb_names = ['building_windows_float_processed', 'building_windows_non_float_processed', 'vehicle_windows_float_processed', 'containers', 'tableware', 'headlamps']
graphviz.Source(tree.export_graphviz(dt1, out_file=None, feature_names=ftr_names, class_names=lb_names))
```

```python
# test the tree on train, get accuracy
acrcy_train = dt1.score(X_train, y_train)

# test the tree on test, get accuracy
acrcy_test = dt1.score(X_test, y_test)

print(acrcy_train, acrcy_test)
```

```python
# build and train a decision tree with most depth = 2
dt2 = DecisionTreeClassifier(max_depth=2)
dt2 = dt2.fit(X_train, y_train)

# visualize the tree
graphviz.Source(tree.export_graphviz(dt2, out_file=None, feature_names=ftr_names, class_names=lb_names))
```

```python
# test the tree on train, get accuracy
acrcy_train = dt2.score(X_train, y_train)

# test the tree on test, get accuracy
acrcy_test = dt2.score(X_test, y_test)

print(acrcy_train, acrcy_test)
```

```python
# test the tree on train, get accuracy
acrcy_train = dt3.score(X_train, y_train)

# test the tree on test, get accuracy
acrcy_test = dt3.score(X_test, y_test)

print(acrcy_train, acrcy_test)
```

# Results Discussion:

1. The trees are very easy to understand, and also make sense. For example,
 
 In Tree1 (max_depth=1), each node has a feature discriminant condition (Ba<=0.335), gini coefficient (gini=0.7465), number of satisfied samples (samples=143), distribution/probability of the samples (value=[48, 47, 11, 10, 7, 20]), and the final decision (class=building_windows_float_processed). This model takes the feature 'Ba' as the most important feature. If the Ba content of a glass is no more than 0.335, then we will classify it as building building_windows_float_processed glass, otherwise a headlamps.
 
 In Tree2 (max_depth=2), this model takes the feature 'Ba' as the first important feature, and takes 'Al' and 'Si' as the second important features.
 
 In Tree3 (max_depth=3), this model takes the feature 'Ba' as the first important feature, 'Al' as the second important feature, 'Ca' and 'Mg' as the third important features. 

2. Generally, the prediction accuracy gets remarkably improved with an increased number of tree's max depth, which indicates that the "deeper" the tree, the better the performance. 
 
 The accuracy of testing on train set gets always higher with max depth increasing. However, the accuracy of testing on test set gets worse or even after max depth >= 4. I think it is due to the overfitting problem. After all, the simple overhigh accuracy on training set is useless. 

3. The accuracy of testing on test set is not as high as I expected. The decision results are also limited. Not all the glass type are detected. I think this may be because the number of instances is still not large enough, and thus the model is not powerful even if it gets max depth of 6.
